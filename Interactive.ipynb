```python
axis_map = {
    "X-coord": 'MappedFixationPointX',
    "Y-coord": 'MappedFixationPointY',
    "Map color": "description",
    "Map name": "StimuliName",
    "Fixation": "FixationDuration",
    "User": "user",
}
```

```python
min_MappedFixationPointY = Slider(title="Minimum Y-coordinates", start=-1662, end=1675, value=-1662, step=1)
min_MappedFixationPointX = Slider(title="Minimum X-coordinates", start=-1264, end=4175, value=-1264, step=1)
max_MappedFixationPointY = Slider(title="Maximum Y-coordinates", start=-1662, end=1675, value=1675, step=1)
max_MappedFixationPointX = Slider(title="Maximum X-coordinates", start=-1264, end=4175, value=4175, step=1)
min_FixationDuration = Slider(title="Minimum FixationDuration", start=16, end=12825, value=0, step=1)
max_FixationDuration = Slider(title="Maximum FixationDuration", start=16, end=12825, value=1000, step=1)
stimulimap = Select(title="Stimulimap", value="All", options=open('Stimulimap.txt').read().split())
user_select = TextInput(title="Map name contains")
desc_select = TextInput(title="Description name contains")
x_axis = Select(title="X Axis", options=sorted(axis_map.keys()), value="X-coord")
y_axis = Select(title="Y Axis", options=sorted(axis_map.keys()), value="Y-coord")

desc = Div(text=open("description.html").read(), sizing_mode="scale_width")
axis_map[x_axis.value]
```

```python
source = ColumnDataSource(data=dict(x=[], y=[], color=[], title=[], user = [], gradient = []))
```

```python
TOOLTIPS=[
    ("Title", "@title"),
    ("Year", "@user"),
    ("X-coordinate", "@x"),
    ("Y-coordinate", "@y"),
]
```

```python
mapper = LinearColorMapper(palette="Magma256", low=33, high=1500, low_color="blue", high_color="red")
```

```python

p = figure(plot_height=600, plot_width=700, title="", tooltips= TOOLTIPS , sizing_mode="scale_both")
p.circle(x="x", y="y", source=source, size=7, fill_color = transform("gradient", mapper), line_color=None)
p.y_range.flipped = True
```

```python
show(p)
```

```python
def select_graphs():
    stimulimap_val = stimulimap.value
    userselect_val = user_select.value.strip()
    desc_val= desc_select.value.strip()
    stimuli_x_smaller = Eyetracking_data['MappedFixationPointX'] <= max_MappedFixationPointX.value
    stimuli_y_smaller = Eyetracking_data['MappedFixationPointY'] <= max_MappedFixationPointY.value
    stimuli_x_greater = Eyetracking_data['MappedFixationPointX'] >= min_MappedFixationPointX.value
    stimuli_y_greater = Eyetracking_data['MappedFixationPointY'] >= min_MappedFixationPointY.value 
    fixation_greater = Eyetracking_data['FixationDuration'] >= min_FixationDuration.value
    fixation_smaller = Eyetracking_data['FixationDuration'] <= max_FixationDuration.value 
   
    
    selected = Eyetracking_data[stimuli_x_greater & stimuli_y_greater & stimuli_x_smaller & stimuli_y_smaller &  
                               fixation_greater & fixation_smaller ]
   
    if (stimulimap_val != "All"):
        selected = selected[selected['StimuliName'].str.contains(stimulimap_val)==True]
    if (userselect_val != ""):
        selected = selected[selected['user'].str.contains(userselect_val)==True]
    if (desc_val != ""):
        selected = selected[selected['description'].str.contains(desc_val)==True]
    return selected
```

```python
def update():
    df = select_graphs()
    x_name = axis_map[x_axis.value]
    y_name = axis_map[y_axis.value]

    p.xaxis.axis_label = x_axis.value
    p.yaxis.axis_label = y_axis.value
    p.title.text = "%d points selected" % len(df)
    source.data = dict(
        x=df[x_name],
        y=df[y_name],
        color=df["description"],
        title=df["StimuliName"],
        user=df["user"],
        gradient = df["FixationDuration"],
    )
```

```python
controls = [min_MappedFixationPointY, max_MappedFixationPointY, min_MappedFixationPointX, max_MappedFixationPointX, min_FixationDuration, max_FixationDuration, stimulimap, user_select, desc_select]
for control in controls:
    control.on_change('value', lambda attr, old, new: update())

inputs = column(*controls, width=320, height=1000)
inputs.sizing_mode = "fixed"
l = layout([
    [desc],
    [inputs, p],
], sizing_mode="scale_both")

update()  # initial load of the data

curdoc().add_root(l)
curdoc().title = "Graph"
```
